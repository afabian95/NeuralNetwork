```python
# Fabian Ardeljan
# Data Mining
# Dr. Ye

from ann import *

training_sets = [
    [[0.6, 0.1], [0, 1]],
    [[0.2, 0.3], [1, 0]]
]

nn = NeuralNetwork(2, 3, 2, 
                   hidden_layer_weights=[0.1, -0.2, 0, 0.2, 0.3, -0.4], 
                   hidden_layer_bias=[0.1, 0.2, 0.5],
                   output_layer_weights=[-0.4, 0.1, 0.6, 0.2, -0.1, -0.2], 
                   output_layer_bias=[-0.1, 0.6],
                   learning_rate=0.1)

print("Original Neural Network:")
nn.inspect()

print("\nIteration: ", 0, "\t\tTotal error: ", round(nn.calculate_total_error(training_sets), 4))
for i in range(10000):
    nn.train(training_sets[0][0], training_sets[0][1])
    nn.train(training_sets[1][0], training_sets[1][1])
    if ((i + 1) % 1000 == 0):
        print("Iteration: ", i + 1, "\tTotal error: ", round(nn.calculate_total_error(training_sets), 4))
        
print("\nFinal Neural Network:")
nn.inspect()
```
